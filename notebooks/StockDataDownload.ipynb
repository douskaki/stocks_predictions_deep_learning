```python
news = [] 

#num_files = 14793
#path1 = '/Users/Dimitris/Downloads/666_20170904105517/666_webhose-2015-07_20170904105917/news_'

def json_to_list(numfiles, path):
    news = []
    for i in range(1,numfiles):
        filename = path + str(i).zfill(7) + '.json'
        with open(filename) as json_file:
            data = json.load(json_file)
            item = []
            item.append(data['published'])
            item.append(data['title'])
            news.append(item)
    return news
```
